```python
# Install the transformers library if not already installed
# !pip install transformers

from transformers import pipeline
import json

# Load the sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Sample 100 comments (duplicated 10 times)
comments = [
    "I love this product! It's amazing.",
    "This is the worst experience I've ever had.",
    "It's okay, not great but not bad either.",
    "Absolutely fantastic service!",
    "Terrible, I will never come back.",
    "The product quality is average.",
    "I'm really happy with the purchase.",
    "Disappointed and frustrated.",
    "Neutral about this.",
    "Exceeded my expectations!"
] * 10  # Total = 100 comments

# Initialize sentiment counters
positive_count = 0
neutral_count = 0
negative_count = 0

# Store labeled results
labeled_comments = []

# Analyze each comment
for comment in comments:
    result = sentiment_pipeline(comment)[0]
    label = result['label']

    if label == "POSITIVE":
        sentiment = "Positive"
        positive_count += 1
    elif label == "NEGATIVE":
        sentiment = "Negative"
        negative_count += 1
    else:
        sentiment = "Neutral"
        neutral_count += 1

    labeled_comments.append({
        "comment": comment,
        "sentiment": sentiment
    })

# Print sentiment summary
print("Sentiment Analysis Summary:")
print("Positive:", positive_count)
print("Neutral:", neutral_count)
print("Negative:", negative_count)

# Save results to a JSON file
with open("labeled_comments.json", "w") as f:
    json.dump(labeled_comments, f, indent=4)

print("Labeled comments saved to 'labeled_comments.json'")
```
